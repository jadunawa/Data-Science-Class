Accurate Average Check

```python
clean_data['Good_Avg'] = clean_data[['TMIN','TMAX','TAVG']].apply(lambda x: True if x[0]<=x[2] and x[2]<=x[1]  else False, axis = 1)
clean_data[clean_data['Good_Avg']==False]
```

Helper Functions that will compute the avg, max, min temperatures 

```python
def avg_temp(row):
    return float((row.TMAX_NEW + row.TMIN_NEW)/2)

def max_temp(row):
    return float(2*row.TAVG-row.TMIN)

def min_temp(row):
    return float(2*row.TAVG-row.TMAX)
```

```python
clean_data['TMIN_NEW'] = clean_data[['TMIN','TMAX','TAVG']].apply(lambda x: min_temp(x) if m.isnan(x[0]) and not m.isnan(x[1]) and not m.isnan(x[2]) else x[0], axis = 1)
clean_data['TMAX_NEW'] = clean_data[['TMIN','TMAX','TAVG']].apply(lambda x: max_temp(x) if m.isnan(x[1]) and not m.isnan(x[0]) and not m.isnan(x[2]) else x[1], axis = 1)
clean_data['TAVG_NEW'] = clean_data[['TMIN_NEW','TMAX_NEW','Good_Avg', 'TAVG']].apply(lambda x: avg_temp(x) if not m.isnan(x[0]) and not m.isnan(x[1]) and x[2]==False else x[3], axis = 1)
```

Here we create a new Series that is indexed by date and contains the TAVG(_NEW) value so we can plot it later

```python
#avg_temp=[]
final_data=[]
for date in clean_data.groupby('DATE').size().index:
    temps_list=clean_data[clean_data['DATE']==date]['TAVG_NEW'].tolist()
    mean=np.mean(temps_list)
    #avg_temp.append(mean)
    row={
        'DATE': date,
        'TAVG': mean
    }
    final_data.append(row)
#print(type(clean_data.groupby('DATE').size().index.tolist()))
```

```python
final_data=pd.DataFrame(final_data)
final_data.set_index('DATE')
```
