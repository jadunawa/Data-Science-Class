```python
def minimum(list):
    minValue=list[0]
    for x in list:
        if(x<minValue):
            minValue=x
    return minValue
```

```python
def maximum(list):
    maxValue=list[0]
    for x in list:
        if(x>maxValue):
            maxValue=x
    return maxValue
```

```python
def mean(list):
    total = 0
    for x in list:
        total = total + x
    meanValue=total/len(list)
    return meanValue
```

```python
def median(lst):
    lst.sort()
    return lst[int(len(lst)/2)]
```

```python
def stdev(list):
    mean_val=mean(list)
    total_difs=0
    for x in list:
        diff=x-mean_val
        difsq=abs(diff)**2
        total_difs=total_difs+difsq
    return (total_difs/len(list))**(1/2)
```
