```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def minimum(list):
    minValue=list[0]
    for x in list:
        if(x<minValue):
            minValue=x
    return minValue
```

```python
def maximum(list):
    maxValue=list[0]
    for x in list:
        if(x>maxValue):
            maxValue=x
    return maxValue
```

```python
def mean(list):
    total = 0
    for x in list:
        total = total + x
    meanValue=total/len(list)
    return meanValue
```

```python
def median(lst):
    lst.sort()
    if len(lst)%2==0:
        return mean([lst[int(len(lst)/2)],lst[int(len(lst)/2)+1]])
    else:
        return lst[int(len(lst)/2)]
```

```python
def stdev(list):
    mean_val=mean(list)
    total_difs=0
    for x in list:
        diff=x-mean_val
        difsq=abs(diff)**2
        total_difs=total_difs+difsq
    return (total_difs/len(list))**(1/2)
```

```python
#get temps as csv
temperatures = pd.read_csv('temperatures.csv')
```

### 1) What were the daily high and low temperatures on the day you were born?

```python
birthday_mask=(temperatures['year']==1995) & (temperatures['month']==12) & (temperatures['day']==13)
birthday_table=temperatures[birthday_mask]
print("Min temp for Judson's birthday: ",birthday_table['tmin'].values[0])
print("Max temp for Judson's birthday: ",birthday_table['tmax'].values[0])
```

### 2) What’s the hottest temperature on record for each month of the year (i.e. the hottest day in January any time in the last 60 years, the hottest day in February, etc.)?

```python
year_mask=(temperatures['year']>temperatures['year'].max()-60)
year_table=temperatures[year_mask]
for i in range(1,13):
    month_mask=(year_table['month']==i)
    month_table=year_table[month_mask]
    print("Max temp for month ",i,": ",maximum(month_table['tmax'].tolist()))
```

### 3) For each day of the year, print the coldest it’s been on that day and the year that record was set

```python
for m in range(1,13):
    for d in range(1,32):
        try:
            #get temp
            day_mask=(temperatures['month']==m) & (temperatures['day']==d)
            day_table=temperatures[day_mask]
            min_temp=minimum(day_table['tmin'].tolist())
            #get year
            year_mask=(day_table['tmin']==min_temp)
            year_table=day_table[year_mask]
            rec_year=year_table['year'].values[0]
            print("Min temp for ",m,"/",d,": ",min_temp,", set in ", rec_year)
        except:
            pass
```

### 4) What’s the earliest day of the year it’s safe to plant frost-sensitive garden plants?

```python
def last_frost(year):
    idk=temperatures[(temperatures['year']==year)]
    mask=((idk['tmin']<33))
    print(mask)
```
