```python
#clean = data.dropna()
clean = data.drop(['Cabin', 'PassengerId', 'Ticket', 'Name'], axis = 1)
meanAge = data.Age.mean()
clean.fillna(meanAge, inplace = True)
clean.info()
```

```python
pSurvived = data.Survived.sum() / data.Survived.count()
print('liklihood of survival:', pSurvived)
```

```python
pSurvived = clean.Survived.sum() / clean.Survived.count()
print('liklihood of survival (cleaned):', pSurvived)
```
