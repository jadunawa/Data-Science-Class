```python
def minimum(list):
    minValue=list[0]
    for x in list:
        if(x<minValue):
            minValue=x
    return minValue
```

```python
def maximum(list):
    maxValue=list[0]
    for x in list:
        if(x>maxValue):
            maxValue=x
    return maxValue
```

```python
def mean(list):
    total = 0
    for x in list:
        total = total + x
    meanValue=total/len(list)
    return meanValue
```

```python
def median(lst):
    lst.sort()
    if len(lst)%2==0:
        return mean([lst[int(len(lst)/2)],lst[int(len(lst)/2)+1]])
    else:
        return lst[int(len(lst)/2)]
```

```python
def stdev(list):
    mean_val=mean(list)
    total_difs=0
    for x in list:
        diff=x-mean_val
        difsq=abs(diff)**2
        total_difs=total_difs+difsq
    return (total_difs/len(list))**(1/2)
```

### 4) What’s the earliest day of the year it’s safe to plant frost-sensitive garden plants?

#### Our temperature threshold is 32 F (freezing point of water)

```python
def streak_days(ar, streak_type, threshold):
    current_start_index=0
    current_streak_length=0
    longest_streak_index=0
    longest_streak_length=0
    current_index=0
    for i in ar:
        #if less than threshold%, increment current streak
        #check if current streak > longest streak, and if so, update longest streak and longest streak start index
        #if more than threshold%, current streak = 0, current start index+=1
        if(i<threshold):
            current_streak_length=current_streak_length+1
            if(current_streak_length>longest_streak_length):
                longest_streak_length=current_streak_length
                longest_streak_index=current_start_index
        else:
            current_start_index=current_index
            current_streak_length=0
        current_index=current_index+1
    #print("Length of",streak_type,": ",longest_streak_length)
    print("Starts at day: ",index_to_day(longest_streak_index+1))
    print("Ends at day: ",index_to_day(longest_streak_index+1+longest_streak_length))
```

```python
freezing_percents = []
for m in range(1,13):
    for d in range(1,32):
        try:
            #get temp
            day_mask=(temperatures['month']==m) & (temperatures['day']==d)
            day_table=temperatures[day_mask]
            freez_num = len(day_table[day_table['tmin']<33].index)
            freezing_percents.append(freez_num/len(day_table.index)*100)
        except:
            pass
streak_days(freezing_percents, "safe planting weather",10)
```

```python
#convert day index to actual date
from datetime import date
def index_to_day(i):
    d=date.fromordinal(i)
    new_format="%m-%d"
    return d.strftime(new_format)
```

### 6) What about summer clothes (e.g. shorts, t-shirts, sandals, etc.)?

#### We assume that summer clothes are necessary if the temperature goes above 70 F.

```python
warm_days = []
for m in range(1,13):
    for d in range(1,32):
        try:
            #get temp
            day_mask=(temperatures['month']==m) & (temperatures['day']==d)
            day_table=temperatures[day_mask]
            warm_num = len(day_table[day_table['tmax']<70].index)
            warm_days.append(warm_num/len(day_table.index)*100)
        except:
            pass
streak_days(warm_days, "warm weather", 25)
```

```python
plt.figure()
plt.plot(warm_days)
plt.title("Warm Day Probability")
plt.xlabel('Days in a Year')
plt.ylabel('Probability (%)')
plt.axhline(y=25, color='r', linestyle='-')
plt.show()
```
